```python
# Finding which years are leap years
leap_years = []
type(leap_years)
for year in range(1900, 2000+1):
    # If year is a century:
    if year % 100 == 0:
        #print(str(year) + " is a century")
        # If century year is not divisible by 400 (not a leap year)
        if year % 400 == 0:
            #print("Century year " + str(year) + " is a leap year")  
            leap_years.append(year)
        elif year % 400 != 0:
            #print("Century year " + str(year) + " is not a leap year")
            pass
    elif year % 4 == 0:
        #print(str(year) + " is a leap year!")
        leap_years.append(year)
    else:
        pass
    
print(leap_years)
```
